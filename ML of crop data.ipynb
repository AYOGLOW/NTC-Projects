#### Column separation

```python
X = df.drop('label', axis=1)  #axis =1 is checking for column, axis - 0 is for row 
y = df['label']  # y is dependent, x= independent
```

#### Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=21, test_size=.25, stratify=y)  #stratify used to maintain 
                                                                                                    # balance in distribution
```

```python
print(X_train.shape)  #shows rows and columns 2325 rows for X and 4 columns.  775 is 25%
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Algorithms

```python
from sklearn.linear_model import LogisticRegression   ## doing classification not linear regression so logistic classifying based on different conditions
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = RandomForestClassifier(random_state=20)  #lr is the algorithm we want to train
lr.fit(X_train,y_train)  #when training it provides both features and target
lr_pred = lr.predict(X_test)  # predict with test saves as lr pred
```

```python
lr_pred
```

### Evaluation

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report
```

```python
print(accuracy_score(y_test,lr_pred))
```

```python
print(classification_report(y_test, lr_pred))
```
